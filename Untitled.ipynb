```python
# Changing the columns for nominal values (Gender (3), Ethnicity (4), Employment Status (4))
```

```python
gender_map = {'Female': 0, 'Male': 1, 'Other': 2}
ethnicity_map = {'Hispanic': 0, 'White': 1, 'Asian': 2, "Black": 3, "Other": 4}
employment_map = {'Employed': 0, 'Retired': 1, 'Student': 3, 'Unemployed': 4}
```

```python
train['gender'] = train['gender'].map(gender_map)
test['gender'] = test['gender'].map(gender_map)
train['ethnicity'] = train['ethnicity'].map(ethnicity_map)
test['ethnicity'] = test['ethnicity'].map(ethnicity_map)
train['employment_status'] = train['employment_status'].map(employment_map)
test['employment_status'] = test['employment_status'].map(employment_map)
```

```python
# Changing the columns for ordinal values (Education level (4), Income level (5), Smoking status (3))
```

```python
education_level_map = {'No formal':0, 'Highschool':1, 'Graduate':2, 'Postgraduate':3}
income_level_map = {'Low':0, 'Lower-Middle':1, 'Middle': 2, 'Upper-Middle': 3, 'High': 4}
smoking_status_map = {'Current': 2, 'Former': 1, 'Never': 0}
```

```python
train['education_level'] = train['education_level'].map(education_level_map)
test['education_level'] = test['education_level'].map(education_level_map)
train['income_level'] = train['income_level'].map(income_level_map)
test['income_level'] = test['income_level'].map(income_level_map)
train ['smoking_status'] = train['smoking_status'].map(smoking_status_map)
test['smoking_status'] = test['smoking_status'].map(smoking_status_map)
```

```python
# Making the values nominal
```

```python
nominal_cols = ['gender', 'ethnicity', 'employment_status']
train = pd.get_dummies(train, columns=nominal_cols)
test = pd.get_dummies(test, columns=nominal_cols)
```

```python
train['diagnosed_diabetes'] = train['diagnosed_diabetes'].astype('int64')
```

```python
# Preparing training variables
```

```python
X_train = train.drop(columns=['diagnosed_diabetes'])
y_train = train['diagnosed_diabetes']
X_test = test
```

```python
# Making the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=1250, random_state=42, min_samples_leaf = 5, min_samples_split = 10, max_features=0.75, max_depth = 100, n_jobs=-1)
```

```python
model.fit(X_train,y_train)
```

```python
submission = pd.read_csv("sample_submission.csv")
```

```python
probabilities = model.predict_proba(X_test)[:, 1]
```

```python
submission['diagnosed_diabetes'] = probabilities
```
